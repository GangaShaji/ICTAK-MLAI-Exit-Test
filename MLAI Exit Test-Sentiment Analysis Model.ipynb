# **Text Preprocessing**

**Step 1: Removing Punctuation marks**

```python
import string
```

```python
string.punctuation
```

```python
punctuationfree=''
def remove_punc(text):
  punctuationfree=''.join([i for i in text if i not in string.punctuation])
  return punctuationfree
```

```python
data['review']=data['review'].apply(lambda x:remove_punc(x))
```

```python
data
```

**Step2: Coverting into lower case**

```python
data['review']=data['review'].apply(lambda x:x.lower())
```

```python
data
```

**Step3: Word Tokenization**

```python
import nltk
nltk.download('punkt')
def tokenize(text):
  word=nltk.word_tokenize(text)
  return word
```

```python
data['review']=data['review'].apply(lambda x:tokenize(x))
```

```python
data.head()
```

**Step4: Removal of stopwords**

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
  output=[i for i in text if i not in stopwords]
  return output
```

```python
data['review']=data['review'].apply(lambda x:remove_stopwords(x))
```

```python
data.head()
```

**Step5:Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
def lemma(text):
  lemm_text=[lemmatizer.lemmatize(x) for x in text]
  return lemm_text
```

```python
data['review']=data['review'].apply(lambda x:lemma(x))
```

```python
data
```

# **Sentiment Analysis Model- using Word2Vec**

**Separating independent and dependent variables**

```python
x=data.review
y=data.sentiment
```

**Train test split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**Training Word2vec**

```python
import gensim
```

```python
w2v_model=gensim.models.Word2Vec(x_train,min_count=1)
```

```python
words=w2v_model.wv.index_to_key
```

```python
x_train_vec = np.array([np.array([w2v_model.wv[i] for i in ls if i in words])for ls in x_train])
x_test_vec = np.array([np.array([w2v_model.wv[i] for i in ls if i in words])for ls in x_test])
```

```python
x_train_vec_avg = []
for v in x_train_vec:
  if v.size:
    x_train_vec_avg.append(v.mean(axis=0))
  else:
    x_train_vec_avg.append(np.zeros(100,dtype = float))

x_test_vec_avg = []
for v in x_test_vec:
  if v.size:
    x_test_vec_avg.append(v.mean(axis=0))
  else:
    x_test_vec_avg.append(np.zeros(100,dtype = float))
```

**Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf=RandomForestClassifier()
rf_clf.fit(x_train_vec_avg,y_train)
```

**Model Evaluation**

```python
y_predictions=rf_clf.predict(x_test_vec_avg)
accuracy_score(y_test,y_predictions)
```

# **Sentiment Analysis Model using ANN**

**Tokenize the text data**

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing import sequence
```

```python
max_words = 10000  # Max vocabulary size
max_sequence_length = 100  # Max sequence length
embedding_dim = 100  # Embedding dimensions
```

```python
tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(data['review'])
```

**Convert tokenized text to sequences**

```python
data['review'] = tokenizer.texts_to_sequences(data['review'])
```

```python
data
```

**Separating dependent and independent variables**

```python
x=data.review
y=data.sentiment
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**Padding sequences to make them equal in length**

```python
x_train=sequence.pad_sequences(x_train,maxlen=max_sequence_length)
x_test=sequence.pad_sequences(x_test,maxlen=max_sequence_length)
```

**ANN Model**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense
```

```python
model = Sequential()
model.add(Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_sequence_length))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=20)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
